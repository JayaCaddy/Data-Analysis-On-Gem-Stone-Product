### Pearson correlation table

```python

import numpy as np


if isinstance(df, (pd.DatetimeIndex, pd.MultiIndex)):
	df = df.to_frame(index=False)

# remove any pre-existing indices for ease of use in the D-Tale code, but this is not required
df = df.reset_index().drop('index', axis=1, errors='ignore')
df.columns = [str(c) for c in df.columns]  # update columns to strings in case they are numbers

corr_cols = [
	 'carat', 'depth', 'table', 'x', 'y', 'z',
	'price'
]
corr_data = df[corr_cols]
corr_data = np.corrcoef(corr_data.values, rowvar=False)
corr_data = pd.DataFrame(corr_data, columns=[corr_cols], index=[corr_cols])
corr_data.index.name = str('column')
corr_data = corr_data.reset_index()
```

```python
corr_data
```

#### From the above PEARSON CORRELATIONS table its clear to interpret that:
     
    -->As Carat of the cubic zirconia is highly correlated with length(x),width(y),height(h) and price of it, Carat of the gem  stone is an essential aspect while decision is made by customers with respect to dimension and price of the same.As           from the univariate analysis its shows that there is lot manufacturing happening in the carat range of 0.4 to 1.05, to         maximize the profit and income to the business: *company has to increase the carat value with more manufactured units*.
     
    --->Depth is slightly negatively correlated with price and as most were manufactured with higher value of depth with the    frequency of 9000 on depth of 62 and 7000 on depth of 63, the company should find ways to make gem stones with less depth as possible to decrease the loss and move towards profit gain.
    
    --->Table values are very slightly correlated with all the other features and does not add much to our analysis.
    
    ---> Length(x),breadth(y) and Height(z) varies positive linearly with price of the cubic zirconia, so to maximize the profit : *Company should start manufacturing more gem stones with bigger dimensions as much as possible*.
    
    ---> Price can be increased with increase in carat and dimensions of the cubic zirconia.
